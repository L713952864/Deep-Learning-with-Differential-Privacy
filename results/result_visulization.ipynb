```python
import matplotlib.pyplot as plt
```

```python
def process_data(file_name):
    file = open(file_name, "r")
    epsilon_raw = file.read()
    spilt_index = epsilon_raw.index('validation')
    eps_str = epsilon_raw[6:spilt_index-2].split(', ')
    accuracy_str = epsilon_raw[spilt_index+17:len(epsilon_raw)-2].split(', ')
    eps = []
    accuracy = []
    for item in eps_str:
        eps.append(float(item))
    for acc in accuracy_str:
        accuracy.append(float(acc))
    return eps,accuracy
```

```python
eps_1,accuracy_1 = process_data("delta_0.01_lr_0.001.txt")
eps_2,accuracy_2 = process_data("delta_0.001_lr_0.001.txt")
eps_3,accuracy_3 = process_data("delta_0.0001_lr_0.001.txt")
eps_4,accuracy_4 = process_data("delta_1e-05_lr_0.001.txt")
plt.plot(eps_1, accuracy_1, label='delta=1e-02')
plt.plot(eps_2, accuracy_2, label='delta=1e-03')
plt.plot(eps_3, accuracy_3, label='delta=1e-04')
plt.plot(eps_4, accuracy_4, label='delta=1e-05')
plt.legend()
plt.xlabel('epsilon')
plt.ylabel('accuracy')
plt.title('Accuracy of various (ε, δ) privacy values')
```
